```python
@app.callback(
    Output(component_id='output-container', component_property='children'),
    [
        Input(component_id='dropdown-statistics', component_property='value'),
        Input(component_id='select-year', component_property='value')
    ]
)
def update_output_container(report_type, selected_year):
    if report_type == 'Recession Period Statistics':
        # Filter data for recession periods (Recession == 1)
        recession_data = df[df['Recession'] == 1]

        # Create 4 plots for the recession data
        fig1 = px.line(recession_data, x='Year', y='Automobile_Sales', title="Automobile Sales during Recession")
        fig2 = px.bar(recession_data, x='Year', y='Advertising_Expenditure', title="Advertising Expenditure during Recession")
        fig3 = px.scatter(recession_data, x='Consumer_Confidence', y='Automobile_Sales', title="Consumer Confidence vs Automobile Sales")
        fig4 = px.box(recession_data, x='Year', y='Price', title="Vehicle Price during Recession")

        # Return the plots as a list inside the output container
        return [
            dcc.Graph(figure=fig1),
            dcc.Graph(figure=fig2),
            dcc.Graph(figure=fig3),
            dcc.Graph(figure=fig4)
        ]

    elif report_type == 'Yearly Statistics':
        # Filter data for the selected year
        yearly_data = df[df['Year'] == int(selected_year)]

        # Create 4 plots for the selected year
        fig1 = px.line(yearly_data, x='Month', y='Automobile_Sales', title=f"Automobile Sales in {selected_year}")
        fig2 = px.bar(yearly_data, x='Month', y='Advertising_Expenditure', title=f"Advertising Expenditure in {selected_year}")
        fig3 = px.scatter(yearly_data, x='Consumer_Confidence', y='Automobile_Sales', title=f"Consumer Confidence vs Automobile Sales in {selected_year}")
        fig4 = px.box(yearly_data, x='Month', y='Price', title=f"Vehicle Price in {selected_year}")

        # Return the plots as a list inside the output container
        return [
            dcc.Graph(figure=fig1),
            dcc.Graph(figure=fig2),
            dcc.Graph(figure=fig3),
            dcc.Graph(figure=fig4)
        ]
```

```python
from dash import Dash, html, dcc, Input, Output
import plotly.express as px

# Initialize the Dash app
app = Dash(__name__)

# Filter data for recession period
recession_data = df[df['Recession'] == 1]

@app.callback(
    Output('output-container', 'children'),
    [Input('dropdown-statistics', 'value'), Input('select-year', 'value')]
)
def update_output_container(report_type, selected_year):
    # Plot 1: Automobile Sales Fluctuate Over Recession Period (Year-wise) using Line Chart
    yearly_rec = recession_data.groupby('Year')['Automobile_Sales'].mean().reset_index()
    R_chart1 = dcc.Graph(
        figure=px.line(yearly_rec, 
            x='Year', 
            y='Automobile_Sales', 
            title="Automobile Sales Fluctuate Over Recession Period (Year-wise)")
    )

    # Plot 2: Average Number of Vehicles Sold by Vehicle Type (Bar Chart)
    average_sales = recession_data.groupby('Vehicle_Type')['Automobile_Sales'].mean().reset_index()
    R_chart2 = dcc.Graph(
        figure=px.bar(average_sales, 
            x='Vehicle_Type', 
            y='Automobile_Sales', 
            title="Average Number of Vehicles Sold by Vehicle Type during Recession")
    )

    # Plot 3: Pie Chart for Total Expenditure Share by Vehicle Type during Recession
    exp_rec = recession_data.groupby('Vehicle_Type')['Advertising_Expenditure'].sum().reset_index()
    R_chart3 = dcc.Graph(
        figure=px.pie(exp_rec, 
            values='Advertising_Expenditure', 
            names='Vehicle_Type', 
            title="Total Expenditure Share by Vehicle Type during Recession")
    )

    # Plot 4: Effect of Unemployment Rate on Vehicle Type and Sales (Bar Chart)
    unemp_data = recession_data.groupby(['unemployment_rate', 'Vehicle_Type'])['Automobile_Sales'].mean().reset_index()
    R_chart4 = dcc.Graph(
        figure=px.bar(unemp_data,
            x='unemployment_rate',
            y='Automobile_Sales',
            color='Vehicle_Type',
            labels={'unemployment_rate': 'Unemployment Rate', 'Automobile_Sales': 'Average Automobile Sales'},
            title='Effect of Unemployment Rate on Vehicle Type and Sales')
    )

    # Returning all the charts inside the output container
    return [
        html.Div(className='chart-item', children=[html.Div(children=R_chart1), html.Div(children=R_chart2)], style={'display': 'flex'}),
        html.Div(className='chart-item', children=[html.Div(children=R_chart3), html.Div(children=R_chart4)], style={'display': 'flex'})
    ]

# Define the layout of the app (add your dropdowns and output-container here)
app.layout = html.Div([
    dcc.Dropdown(id='dropdown-statistics', options=[{'label': 'Yearly Statistics', 'value': 'Yearly Statistics'}, 
                                                    {'label': 'Recession Period Statistics', 'value': 'Recession Period Statistics'}], 
                 placeholder='Select a report type'),
    dcc.Dropdown(id='select-year', options=[{'label': str(i), 'value': str(i)} for i in range(1980, 2024)], 
                 placeholder='Select a year'),
    html.Div(id='output-container', className='chart-grid')
])

if __name__ == '__main__':
    app.run_server(debug=True)
```

```python
@app.callback(
    Output('output-container', 'children'),
    [Input('dropdown-statistics', 'value'), Input('select-year', 'value')]
)
def update_output_container(selected_statistics, input_year):
    # If 'Yearly Statistics' is selected and a year is chosen
    if selected_statistics == 'Yearly Statistics' and input_year:
        # Filter data for the selected year
        yearly_data = df[df['Year'] == int(input_year)]

        # Plot 1: Yearly Automobile Sales using Line Chart
        yas = df.groupby('Year')['Automobile_Sales'].mean().reset_index()  # Group by year to get the average sales
        Y_chart1 = dcc.Graph(
            figure=px.line(yas, 
                x='Year', 
                y='Automobile_Sales', 
                title=f"Yearly Automobile Sales for {input_year}")
        )

        # Plot 2: Total Monthly Automobile Sales using Line Chart
        mas = df.groupby('Month')['Automobile_Sales'].sum().reset_index()  # Group by month to get total sales per month
        Y_chart2 = dcc.Graph(
            figure=px.line(mas, 
                x='Month', 
                y='Automobile_Sales', 
                title="Total Monthly Automobile Sales")
        )

        # Plot 3: Average Number of Vehicles Sold by Vehicle Type (Bar Chart)
        avr_vdata = yearly_data.groupby('Vehicle_Type')['Automobile_Sales'].mean().reset_index()
        Y_chart3 = dcc.Graph(
            figure=px.bar(avr_vdata, 
                x='Vehicle_Type', 
                y='Automobile_Sales', 
                title=f"Average Vehicles Sold by Vehicle Type in {input_year}")
        )

        # Plot 4: Total Advertisement Expenditure for Each Vehicle (Pie Chart)
        exp_data = yearly_data.groupby('Vehicle_Type')['Advertising_Expenditure'].sum().reset_index()
        Y_chart4 = dcc.Graph(
            figure=px.pie(exp_data, 
                values='Advertising_Expenditure', 
                names='Vehicle_Type', 
                title="Total Advertisement Expenditure for Each Vehicle")
        )

        # Return all charts inside divs
        return [
            html.Div(className='chart-item', children=[html.Div(children=Y_chart1), html.Div(children=Y_chart2)], style={'display': 'flex'}),
            html.Div(className='chart-item', children=[html.Div(children=Y_chart3), html.Div(children=Y_chart4)], style={'display': 'flex'})
        ]
    # Return empty if 'Recession Period Statistics' is selected (or no year is selected)
    return []

```
